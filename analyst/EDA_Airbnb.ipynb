```python
import pandas as pd

```

```python
df = pd.read_csv("../data/Airbnb+Data/Airbnb Data/Listings.csv", engine='pyarrow', encoding='iso-8859-1')
```

```python
df = df[df['host_location'].str.contains("United States")]
df = df[df['host_location'] != "Paris, France/Seattle, Washington, United States"]

# Dividir la columna 'host_location' en 'barrio', 'estado' y 'país'
split_columns = df['host_location'].str.split(', ', expand=True)

# Rellenar las columnas faltantes con valores vacíos para asegurar que siempre haya 3 columnas
while split_columns.shape[1] < 3:
    split_columns[split_columns.shape[1]] = ''

# Asignar los nombres de las nuevas columnas
split_columns.columns = ['city', 'state', 'country']

df.rename(columns={'host_location': 'country', 'city': 'state'}, inplace=True)

# Combinar las nuevas columnas con el DataFrame original
df[['district', 'state', 'country']] = split_columns


df["country"] = "United States"
# reordering columns
cols = ["listing_id", "name", "country", "state", "district", "latitude", "longitude", "property_type", "room_type", "accommodates", "amenities", "price", "minimum_nights", "maximum_nights", "review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value", "host_response_time", "host_response_rate", "host_acceptance_rate", "host_total_listings_count", "host_identity_verified", "host_has_profile_pic", "instant_bookable"]
df = df[cols]
```
